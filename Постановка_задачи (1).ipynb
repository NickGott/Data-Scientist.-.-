##### 
В целом данные корректны, но встречаются двузначные ответы больше 10 (видимо пропущены запятые). Обработаем столбец и уберем некорректные данные.

```python
#  Функция по разметке ответов Q2.
# mark - строка с ответами на вопрос. Функция возвращает список цифр диапазона ответа без повторов
def q2_map(mark):
    # Регулярным выражением делим строку на список чисел
    ex = re.findall('\w+', mark)

    # Пробразуем к целому
    ex = list(map(int, ex))

    # Новый список для записи
    new_list = []

    # Итерируемся по списку и проверяем на соответстие условию
    for element in ex:
        if 1 <= element <= 7:
            new_list.append(element)  # Если верный диапазон отметки
        elif element > 10:
            # Если двузначное число, делим по символам и проверяем вхождение в диапазон
            new_list += [int(x) for x in str(element) if 1 <= int(x) <= 7]
        else:
            pass
    # На выходе удаляем дубликаты
    return list(set(new_list))
```

```python
df.fillna('0', inplace=True)
# Разбиение строки в столбце на список значений(ответов) на Q2
df['Q2'] = df['Q2'].apply(lambda x: q2_map(x))

# Количество претензионных оценок на клиента
df['DisCount'] = df['Q2'].apply(lambda x: len(x))
```

### Индекс удовлетворенности клиентов сервисом - CSAT.
Рассчитаем CSAT – Customer Satisfaction Score

CSAT – это Индекс удовлетворённости клиента, эта метрика призвана замерять уровень "клиентского счастья".
Иными словами это некая средняя величина, выражающая степень соответствия реальных услуг и товаров тем, которые клиенты должны были предположительно получить.
В нашем случае оценка ведется по десятибальной шкале: от "ужасно" к "отлично".

Зачем измерять?
Оценка удовлетворенности требуется для сбора обратной связи клиентов с помощью опросов. Чтобы знать наверняка мнение пользователя, нужно просто спросить, каковы его впечатления от продукта или сервиса.



Какие недостатки у метода CSAT?

- Возможно искажение результатов. Недовольные клиенты склонны в принципе не проходить опросы (исключение — очень недовольные клиенты, которые используют любую возможность, чтобы сообщить о своём опыте). Пользователи, оценка которых колеблется от нейтральной до неудовлетворительной, часто не отвечают на вопросы. Довольные же клиенты охотно соглашаются из соображений благодарности.

- Субъективность и неоднозначность. У разных клиентов разные ожидания. Для кого-то лишняя фраза, сказанная менеджером, может стать пыткой и говорить об ужасном сервисе, а кому-то важны любезности при разговоре.

##### Расчет CSAT будем строить так: CSAT Score (%) = (Sum of all scores) / (Sum max. Possible score) * 100

Также ведем градацию оценок:

- 1-2 балла - ужасно;
- 3-4 балла - плохо;
- 5-6 баллов - нейтрально;
- 7-8 баллов - хорошо;
- 9-10 баллов - отлично.

##### На основании полученных оценок Q1 рассмотрим показатели:

- Уровень удовлетворенности клиентов предоставляемыми услугами CSAT;
- Средняя оценка уровню сервиса.

```python
evaluation = {1: 'ужасно',
              2: 'ужасно',
              3: 'плохо',
              4: 'плохо',
              5: 'нейтрально',
              6: 'нейтрально',
              7: 'хорошо',
              8: 'хорошо',
              9: 'отлично',
              10: 'отлично'}

df['evaluation'] = df['Q1'].map(evaluation)
```

```python
index = df['Q1'].sum() / (df['Q1'].count() * 10)
mean_csat = df['Q1'].sum() / df['Q1'].count()

print(f'Процент пользователей, удовлетворенных услугами - {index:.2%}')
print(f'Средняя оценка, удовлетворенности услугами - {mean_csat:.2}')
```

Мы нашли процент клиентов, удовлетворенных качеством сервиса. Но мы имеем 10-ти бальную шкалу и интересно посмотреть как распределились оценки участников опроса при ответе на первый вопрос.

```python
def custom_hist(data, col, title=None, xlab=None, ylab=None, median=False, mean=False, hue=None):
    """Функция отрисовки гистограммы"""

    plt.figure(figsize=(10, 5))
    sns.histplot(data, x=col, kde=False, hue=hue, discrete=True)
    plt.title(title, fontsize=14, pad=15)
    plt.ylabel(ylab)
    plt.xlabel(xlab)
    plt.xticks(np.arange(data[col].min(), data[col].max()+1))

    if mean is True:
        plt.axvline(data[col].mean(), color='m',
                    linestyle='dashed', label='среднее', alpha=0.5)
        plt.legend()

    if median is True:
        plt.axvline(data[col].median(), color='green',
                    linestyle='dashed', label='медиана', alpha=0.7)
        plt.legend()

    return plt.show()
```

```python
custom_hist(data=df, col='Q1', title='Распределение оценок Q1',
            xlab='Оценки Q1', ylab='Количество пользователей', median=True, mean=True, hue='evaluation')
```

По распределению можно увидеть, что большинство при ответе на вопрос, поставило оценку 10 баллов, в тоже время мы видим большое число людей поставивших оценку "очень плохо". #### Плотность распределения технических параметров связи

Рассмотрим плотность распределения по двум группам:

1. Группа довольных сервисом. Оценка на Q1 больше 6.
2. Группа недовольных сервисом. Оценка на Q1 меньше 5.

Ранее мы уже выявили 3 основные причины заниженных оценок.

- Плохое качество связи в зданиях - Q2 - ответ 3;
- Недозвоны, обрывы при звонках - Q2 - ответ 1;
- Медленный мобильный интернет - Q2 - ответ 4.

```python
def plot_all_group(data, log=False):
    """Функция по визуализации причин по которой клиенты снижали оценку.
    Сначала убирает нейтральные оценки и сравнивает две группы с хорошими оценками и плохими"""
    
    # Отсеиваем оценки
    data = data[(data.Q1 >= 7) | (data.Q1 <= 4)]
    # Делим на две группы
    data['evaluation'] = data['Q1']\
        .apply(lambda x: 'Not satisfied' if x <= 4 else 'Satisfied')
    # Формируем полотно
    fig, axes = plt.subplots(2, 4, figsize=(20, 8), sharey=False)

    fig.suptitle(f'Плотность распределения', y=1., fontsize=14)

    num = 0
     # Заполняем ax
    for i in range(0, 2):
        for j in range(0, 4):
            param = tech_params[num]
            sns.histplot(ax=axes[i][j], data=data[data[param] != 0],
                         x=param, hue='evaluation', stat='density',
                         hue_order=data.evaluation.unique(),
                         kde=True, log_scale=log, legend=False)

            axes[i][j].set_ylabel('Плотность')
            # Вычленяем единицу измерения
            if log == True:
                axes[i][j].set_xlabel('Log: ' + re.findall(r'\((\S+)\)', tech_params[num])[0])
            else:
                axes[i][j].set_xlabel(re.findall(r'\((\S+)\)', tech_params[num])[0])
                
            axes[i][j].set_title(param)
            plt.legend(data.evaluation.unique())
            sns.despine()
            num += 1
    plt.subplots_adjust(wspace=0.25, hspace=.4)
    return plt.show()
```

```python
plot_all_group(data=df, log=False)
```

Видно, что плотность распределения у двух групп несколько различается, также можно отметить длинные хвосты. Визуально оценим теже данные на логарифмированной шкале.

### Проверка гипотез

#### Cредняя скорость «к абоненту» (Downlink Throughput(Kbps)) в двух группах не отличается

Принимается статистическая гипотеза об отсутствии статистической разницы в двух группах по характеристике Downlink Throughput

```python
def bootstrap(data_a, data_b, N=5000, size=200, title=None):
    """Функция бутстрэпа, расчет и отрисовка"""
    
    # Реализация бутстрэпа
    mean_samples_a = []
    mean_samples_b = []
    buf = []
    
    for i in range(N):
        sample_a =  data_a.sample(size)
        mean_samples_a.append(np.mean(sample_a))
        sample_b =  data_b.sample(size)
        mean_samples_b.append(np.mean(sample_b))
        buf.append(np.mean(sample_a)-np.mean(sample_b))
        
    # Инициализация полотна
    fig, ax = plt.subplots(1, 2, figsize=(16, 6))
    fig.suptitle(title, y=1.05, fontsize=14)
    
    # Графики распределения по группам А и Б
    sns.histplot(mean_samples_a, alpha=.8, ax=ax[0], kde=True, stat='density')
    sns.histplot(mean_samples_b, color='orange', alpha=.8, ax=ax[0], kde=True, stat='density')
    # Интервалы и медиана
    ax[0].axvline(np.percentile(mean_samples_a, 2.5), linestyle='-', color='r', linewidth=3)
    ax[0].axvline(np.percentile(mean_samples_a, 97.5), linestyle='-', color='r', linewidth=3)
    ax[0].axvline(np.median(mean_samples_a), linestyle='--', color='r', linewidth=3)
    ax[0].axvline(np.percentile(mean_samples_b, 2.5), linestyle='-', color='g', linewidth=3)
    ax[0].axvline(np.percentile(mean_samples_b, 97.5), linestyle='-', color='g', linewidth=3)
    ax[0].axvline(np.median(mean_samples_b), linestyle='--', color='g', linewidth=3)
    ax[0].set_title('Плотность распределения по группам \n с хорошими и плохими оценками', fontsize=14)
    ax[0].set_ylabel('Плотность')
    ax[0].set_xlabel(re.findall(r'\((\S+)\)', title)[0])
    ax[0].legend(['Хорошие','Плохие'])
    
    # График распределения разности по группам А и Б
    sns.histplot(buf, alpha=.8, ax=ax[1], kde=True, stat='density', color='g')
    # Интервалы и медиана
    ax[1].axvline(np.percentile(buf, 2.5), linestyle='-', color='b', linewidth=3)
    ax[1].axvline(np.percentile(buf, 97.5), linestyle='-', color='b', linewidth=3)
    ax[1].axvline(np.median(buf), linestyle='--', color='b', linewidth=3)
    ax[1].set_title('Плотность разницы средних', fontsize=14)
    ax[1].set_ylabel('Плотность')
    ax[1].set_xlabel(re.findall(r'\((\S+)\)', title)[0])
    return plt.show()
```

```python
bootstrap(data_a=df[df['Q1'] >= 7]['Downlink Throughput(Kbps)'],
          data_b=df[df['Q1'] <= 4]['Downlink Throughput(Kbps)'],
          N=5000, size=300, title='Downlink Throughput(Kbps)')
```

Доверительные интервалы пересекаются, но не накладываются. Медианные значения далеки друг от друга. Разница среднего не проходит через ноль. Нулевую гипотезу равенства отвергаем.
